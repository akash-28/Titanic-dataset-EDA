```python
def func1(c):
    age = c[0]
    pclass = c[1]
    Sex = c[2]
    
    if pd.isnull(age):
        if pclass == 1 and Sex == 'male':
            return 40
        elif pclass == 1 and Sex == 'female':
            return 34
        elif pclass == 2 and Sex == 'male':
            return 30
        elif pclass == 2 and Sex == 'female':
            return 27
        elif pclass == 3 and Sex == 'male':
            return 25
        elif pclass == 3 and Sex == 'female':
            return 22
    else:
        return age
```

## converting categorical data to numerical data


```python
## we use label encoder, onehotencoder (from sklearn)
## get_dummies method present in pandas

df.head()
```

```python
# create dummies for Age, Sex and Embark
```

```python
s_ = pd.get_dummies(df.Sex)
s_.head()

#gives highest number to the first value like male has got value 1 ... in this case

#since both column contains same data...ex: any one of the column also has data for the other column... 
#so we'll drop the first column
#same is true for the other categorical variables
```

```python
s_ = pd.get_dummies(df.Sex, drop_first=True)
s_.head()
```

```python
e_ = pd.get_dummies(df.Embarked, drop_first=True)
e_.head()
```

```python
#now concatenating these varibale to the actual data set
df = pd.concat([df, s_, e_], axis=1)
df.head()
```

```python
df.drop(['Sex', 'Embarked'], axis=1, inplace=True)
```
